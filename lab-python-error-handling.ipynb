````python
# Step 1: Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError(
                        "Invalid quantity! Please enter a non-negative value."
                    )
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory


# 3. Modify the `get_customer_orders` function to include error handling.
def get_customer_orders(inventory):
    customer_orders = set()
    valid_input = False
    while not valid_input:
        try:
            number_of_orders = int(input("Enter the number of customer orders: "))
            if number_of_orders <= 0:
                raise ValueError(
                    "Invalid number of orders! Please enter a valid value."
                )
            else:
                valid_input = True
        except Exception as e:
            print(
                "Invalid number of orders! Please enter a valid numeric value.", str(e)
            )

    valid_enter = False
    while not valid_enter:
        try:
            for _ in range(int(number_of_orders)):
                order = input(
                    f"Enter the name of a product that a customer wants to order: "
                )
                if order not in inventory:
                    raise ValueError("Invalid product! Please enter a valid product.")
                elif inventory[order] == 0:
                    raise ValueError(
                        f"{order} is out of stock. The product has been removed from your order."
                    )
                else:
                    customer_orders.add(order)
                    print(f'"{order}" has been added to your order.')
            valid_enter = True
        except ValueError as e:
            print(e)
            customer_orders.clear()

    return customer_orders


def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory:
            if inventory[product] > 0:
                inventory[product] -= 1
                if inventory[product] == 0:
                    print(f"{product} is now out of stock.")
            else:
                print(
                    f"{product} is out of stock. The product has been removed from your order."
                )
                customer_orders.remove(product)
        else:
            print(f"The product '{product}' is not in the inventory.")

    # Use dictionary comprehension to filter out products with zero quantity
    inventory = {
        product: quantity for product, quantity in inventory.items() if quantity > 0
    }
    return inventory


def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_unique_products = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_unique_products


def print_order_statistics(order_statistics):
    total_products_ordered, percentage_unique_products = order_statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_unique_products:.2f}%")


def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")


# Step 2 Modify the `calculate_total_price` function to include error handling.
def calculate_total_price(customer_orders, inventory):
    total_price = 0
    for product in customer_orders:
        valid_input = False
        while not valid_input:
            try:
                prices = input(f"Enter the price of {product}: ")
                if int(prices) < 0:
                    raise ValueError(
                        "Invalid price! Please enter a non-negative value."
                    )
                elif prices.isdigit() == False:
                    raise ValueError(
                        "Invalid price! Please enter a valid numeric price."
                    )
                else:
                    valid_input = True
            except Exception as e:
                print("Invalid input:", str(e))
        total_price += int(prices)
    print(f"Total price: {total_price}")


# List of available products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Initialize inventory
inventory = initialize_inventory(products)

# Get customer orders
customer_orders = get_customer_orders(inventory)

# Update inventory based on customer orders
update_inventory(customer_orders, inventory)

# Calculate order statistics
order_statistics = calculate_order_statistics(customer_orders, products)

# Print order statistics
print_order_statistics(order_statistics)

# Print updated inventory
print_updated_inventory(inventory)

# Print total price
calculate_total_price(customer_orders, inventory)
````
